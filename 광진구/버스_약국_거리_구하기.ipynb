```python
##거리 계산을 위한 모듈 가져오기
from haversine import haversine
```

```python
min_lst=[]
for i in range(len(pharm)):
    busstop_name=pharm.iloc[i]['약국명']
    busstop=(pharm.iloc[i]['Y좌표'],pharm.iloc[i]['X좌표'])
    min_d=10000000
    #busstop_goo=busline.iloc[i]['행정구']
    
    for j in range(len(busfinal)):
        df=busfinal.iloc[j]
        sub_name=df['정류소명']
        sub=(df['Y좌표'],df['X좌표'])
        distance=haversine(busstop, sub, unit = 'm')
        
        if min_d>distance:
            min_d=distance
            min_sub_name=sub_name
    
    min_dic={
        '약국명':busstop_name,
        '최소 거리 버스정류소': min_sub_name,
        '정류소 최소 거리':min_d
    }
    #print(min_dic)
    #print(dong_name,min_wel_name,min_d)
    min_lst.append(min_dic)
```

```python
# sub_pharm = pd.DataFrame(min_lst)
pd.DataFrame(min_lst)
```

```python
pd.DataFrame(min_lst).to_csv(path + '버스_최소_거리.csv')
```
