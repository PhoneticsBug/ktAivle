#### [실습문제1] 데이터 로딩
* '1.4 business_type.csv'파일을 'seoul_business' 변수에 저장하고 그 데이터를 확인하세요.
    * 데이터 파일 로딩시 참고 사항 
        * 구분자(sep)는 ',' 입니다
        * UTF-8 인코더를 사용해 주세요

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
import pandas as pd

business = pd.read_csv('C:/Users/User/iCloudDrive/AIVLE/3주차/2023.02.13_미니프로젝트1차_실습자료/1.4 business_type.csv', sep=',', encoding='utf-8')
```

```python
# 데이터 프레임의 Shape을 확인합니다.
business.shape
```

```python
business
```

```python
seoul_business = business.iloc[:,[0,1,2,5,13,17, 19]]
seoul_business.head()
```

```python
seoul_business = seoul_business[seoul_business['동'] == '소계'].reset_index(drop=True)
```

```python
seoul_business = seoul_business.drop('동', axis=1)
```

```python
seoul_business
```

```python
seoul_business.rename(columns={"다수업종 상위10": "택시", "다수업종 상위10.3":"한식", "다수업종 상위10.11":"커피", "다수업종 상위10.15":"주점업", "다수업종 상위10.17":"학원업"}, inplace=True)
```

```python
seoul_business.head()
```

```python
# per_capita_1['USD_2'] = per_capita_1.USD.str.replace(',', '').astype('int64')

seoul_business['택시운송업'] = seoul_business.택시.str.replace(',', '').astype('int64')
seoul_business['한식및일반음식점'] = seoul_business.한식.str.replace(',', '').astype('int64')
seoul_business['커피전문점'] = seoul_business.커피.str.replace(',', '').astype('int64')
seoul_business['기타주점업'] = seoul_business.주점업.str.replace(',', '').astype('int64')
seoul_business['학원'] = seoul_business.학원업.str.replace(',', '').astype('int64')
```

```python
seoul_business.head()
```

```python
business_seoul = seoul_business.iloc[:,[0, 6, 7, 8, 9, 10]]
business_seoul
```

```python
business_seoul.to_csv('C:/Users/User/iCloudDrive/AIVLE/3주차/2023.02.13_미니프로젝트1차_실습자료/df_seoul_business.csv', index = False)
```
